```python
state = input("Enter your name: ")
import pandas as pd
df = pd.read_csv("B2C_sales_data.csv")
df.head()

state_group = df.groupby(["Seller Gstin", "Ship To State"])
if state == "ALL":
    state_group = df.groupby(["Seller Gstin"])
    delhi_gujarat = state_group.get_group(('24AVJPJ4767F1ZG'))
else:
    state_group = df.groupby(["Seller Gstin", "Ship To State"])
    delhi_gujarat = state_group.get_group(('24AVJPJ4767F1ZG',state))

sales = delhi_gujarat.loc[:, ['Ship To State', 'Hsn/sac','Igst Rate','Quantity','Transaction Type','Tax Exclusive Gross','Igst Tax','Cgst Rate','Sgst Rate']]
# sales['Igst Rate'] = sales['Igst Rate'].replace(0.05,5)
sales

total = sales.assign(total_tax = sales['Tax Exclusive Gross'] + sales['Igst Rate'] + sales['Cgst Rate'] + sales['Sgst Rate'])
total

total.loc[delhi_gujarat['Transaction Type'].isin(['Refund', 'Cancel']), 'Quantity'] *= -1
total = total.drop(total[total['Tax Exclusive Gross'] == 0].index)
total

all_total = total.groupby(['Ship To State','Hsn/sac', 'Igst Rate']).agg({'Quantity': 'sum', 'Tax Exclusive Gross': 'sum', 'Igst Tax': 'sum','Cgst Rate':'sum','Sgst Rate':'sum', 'total_tax': 'sum'})
all_total


result = all_total.reset_index()
result
```
